# Training with Transfer Learning with YOLO3 Starts

```python
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="Computer_Vision_Intern_assignment")
trainer.setTrainConfig(object_names_array=["crack","bullseye","scratch"], batch_size=2, num_experiments=100, train_from_pretrained_model="pretrained-yolov3.h5")
trainer.trainModel()
```

The best model weights with pretrained YOLO3 is detection_model-ex-085--loss-0002.644.h5 file.

We get the best loss of 0002.644

# Evaluation Stage Starts

```python
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="Computer_Vision_Intern_assignment")
trainer.evaluateModel(model_path="Computer_Vision_Intern_assignment/models", json_path="Computer_Vision_Intern_assignment/json/detection_config.json", iou_threshold=0.4, object_threshold=0.3, nms_threshold=0.4)
```

# Tesing on a Image 

```python
from imageai.Detection.Custom import CustomObjectDetection

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("Computer_Vision_Intern_assignment/models/detection_model-ex-085--loss-0002.644.h5") 
detector.setJsonPath("Computer_Vision_Intern_assignment/json/detection_config.json")
detector.loadModel()
detections = detector.detectObjectsFromImage(input_image="5566.jpg", output_image_path="output_5566_detected.jpg")
for detection in detections:
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
```
